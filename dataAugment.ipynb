```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
pictures = ['10000X_001.tif','10000X_007.tif','10000X_008.tif','10000X_009.tif','10000X_010.tif','10000X_011.tif']
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import layers

data_augmentation = tf.keras.Sequential([
  layers.RandomRotation(0.2),
  layers.RandomFlip("horizontal_and_vertical"),
  
])
```

```python
from PIL import Image

# path to save augmented images
saved_path='/content/drive/My Drive/SURE Summer 2022/Augmented_Bainite/'
plt.figure(figsize=(10, 10))

for j in range(len(pictures)):
  img = cv2.imread(path + pictures[j])
  gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
  for i in range(9):
    # augment image
    augmented_image = data_augmentation(tf.expand_dims(gray_img, 0))
    # save augmented images to drive
    image = Image.fromarray(np.array(augmented_image[0]).astype(np.uint8))
    image.save(saved_path + 'Bainite'+ str(j) + str(i), format='PNG')
    #ax = plt.subplot(3, 3, i + 1)
    #plt.imshow(augmented_image[0], cmap='gray')
    #plt.axis("off")
```
